# Data Analysis

Top 5 most popular artists.


```python
## Grouping the Artist and the song title together to find out which artist has the most songs under their name.

top_five_artists = df.groupby("artist").count().sort_values(by = "song_title", ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()

plt.show()
```

# Top 5 loudest track

```python
to_five_loudest_tracks = df[["loudness", "song_title"]].sort_values(by="loudness", ascending=True)[:5]
to_five_loudest_tracks
```

```python
## Loudness is based on -33.097 > -31.367 > -31.082 
## So now its arranged in descending order.
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="loudness", y="song_title", data=to_five_loudest_tracks)
plt.title("Top 5 loudest Tracks")
plt.show()
```

Artist with Most danceability song

```python
top_five_artist_danceability_songs = df[["danceability", "song_title", "artist"]].sort_values(by="danceability", ascending=False)[:5]
top_five_artist_danceability_songs
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x = "danceability", y = "artist", data=top_five_artist_danceability_songs)
plt.title("Artist With The Most Danceability Song")
plt.show()
```

Top 10 Instrumental tracks

```python
top_ten_instrumental_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12, 7))
plt.pie(x="instrumentalness", data = top_ten_instrumental_tracks, autopct = '%1.2f%%', labels= top_ten_instrumental_tracks.song_title)
plt.show()
```
